```python
def mean_recall_at_10(predictions_df, ground_truth_df):
    recalls = []
    for query in ground_truth_df['query'].unique():
        gt_urls = set(ground_truth_df[ground_truth_df['query'] == query]['assessment_url'])
        pred_urls = set(predictions_df[predictions_df['query'] == query]['url'].head(10))
        recall = len(pred_urls & gt_urls) / len(gt_urls) if gt_urls else 0
        recalls.append(recall)
    return sum(recalls) / len(recalls)
```
